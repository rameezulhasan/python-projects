```python
from random import randint
computer_value = randint(1,100)
a = -1
guesses = 0
while(a != computer_value):
    guesses +=1
    take_player_value = int(input("Enter Any Number From 1 To 100 To Guess Correct Number"))
    if (take_player_value > computer_value):
        print("Enter Lower Value From This")
    elif (take_player_value< computer_value):
        print("Enter Greater Value From This")
    elif (take_player_value == computer_value):
        print(f"Gonratulations! You Entered Correct Number in {guesses} guesses")
        break
    elif (take_player_value > 100):
        print("Please Enter Lower Value From 100")
    elif (take_player_value < 0):
        print("Please Enter Greater Value From 0")
```
